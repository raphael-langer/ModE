# CHP_heat neu

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

def linearize_function(func, num_points, E_out_min, E_out_max):
    # Erzeuge E_in-Werte im großen Bereich (hier nehme ich z.B. 1000 bis 50000)
    E_in = np.linspace(1000, 50000, num_points).reshape((-1, 1))
    
    # Erzeuge E_out-Werte basierend auf der Funktion
    E_out = func(E_in)
    
    # Filtere die Daten basierend auf den E_out-Werten
    mask = (E_out >= E_out_min) & (E_out <= E_out_max)
    E_in_filtered = E_in[mask].reshape((-1, 1))
    E_out_filtered = E_out[mask].reshape((-1, 1))
    
    # Erstelle das lineare Regressionsmodell
    model = LinearRegression()
    model.fit(E_in_filtered, E_out_filtered)
    
    # Vorhersagen
    E_out_pred = model.predict(E_in_filtered)
    
    # Plot der Daten und der Regressionslinie
    plt.scatter(E_in_filtered, E_out_filtered, color='black', label='Gefilterte Originaldaten')
    plt.plot(E_in_filtered, E_out_pred, color='blue', linewidth=3, label=f'Regressionslinie: E_out = {model.coef_[0][0]:.2f} * E_in + {model.intercept_[0]:.2f}')
    plt.xlabel('E_in')
    plt.ylabel('E_out')
    plt.title('Linearisierung der Funktion (Gefilterte Daten)')
    plt.legend()
    plt.grid(True)
    
    # Anzeigen der Koeffizienten und des Intercepts
    print(f"Steigung (Slope): {model.coef_[0][0]}")
    print(f"y-Achsenabschnitt (Intercept): {model.intercept_[0]}")
    
    # Berechnung des mittleren quadratischen Fehlers (MSE)
    mse = mean_squared_error(E_out_filtered, E_out_pred)
    print(f"Mean Squared Error: {mse}")
    
    # Berechnung der E_out-Werte für die tatsächlichen minimalen und maximalen E_in-Werte der gefilterten Daten
    E_in_min_filtered = E_in_filtered.min()
    E_in_max_filtered = E_in_filtered.max()
    E_out_min_val = func(np.array([[E_in_min_filtered]]))[0]
    E_out_max_val = func(np.array([[E_in_max_filtered]]))[0]
    print(f"E_out-Wert für E_in_min ({E_in_min_filtered}): {E_out_min_val}")
    print(f"E_out-Wert für E_in_max ({E_in_max_filtered}): {E_out_max_val}")

# Beispiel-Funktion (kann durch jede andere Funktion ersetzt werden)
def example_function(E_in):
    return 3585 - 4740740.741 * (1 / E_in)

# Parameter für die Analyse
num_points = 1000

# Wertebereich für E_out
E_out_min = 0
E_out_max = 800

# Linearisierung der Beispiel-Funktion
linearize_function(example_function, num_points, E_out_min, E_out_max)
```
