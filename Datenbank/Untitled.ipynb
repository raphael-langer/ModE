```python
import json
import matplotlib.pyplot as plt

# JSON-Datei im aktuellen Arbeitsverzeichnis
file_name = 'GM_Heur_0_5.json'

# JSON-Daten laden
with open(file_name, 'r') as f:
    variable_values = json.load(f)

# Funktion zum Extrahieren von Variablenwerten aus den JSON-Daten
def extract_variable_values(variable_name):
    data = variable_values[variable_name]
    time_steps = sorted(data.keys(), key=lambda x: int(x))  # Zeitpunkte sortieren
    values = [data[ts] for ts in time_steps]
    return time_steps, values
```

```python
import matplotlib.pyplot as plt
import numpy as np
import os

# Beispiel-Daten
components = ['Component 1', 'Component 2', 'Component 3', 'Component 4']
values1 = [10, 20, 15, 25]  # Werte für Gruppe 1
values2 = [12, 18, 17, 22]  # Werte für Gruppe 2

x = np.arange(len(components))  # die Labelpositionen
width = 0.2  # die Breite der Balken (verkleinert, um sie dünner zu machen)

# Sicherstellen, dass der Ordner "Plot pics/Volllaststunden" existiert
base_dir = 'Plot pics'
output_dir = os.path.join(base_dir, 'Volllaststunden')
os.makedirs(output_dir, exist_ok=True)

fig, ax = plt.subplots(figsize=(10, 6))

# Erstellen der Balken für beide Gruppen
rects1 = ax.bar(x - width/2, values1, width, label='Group 1', color='blue')
rects2 = ax.bar(x + width/2, values2, width, label='Group 2', color='orange')

# Hinzufügen von Text für Labels, Titel und Achsen
ax.set_xlabel('Components')
ax.set_ylabel('Values')
ax.set_title('Grouped Bar Chart Example')
ax.set_xticks(x)
ax.set_xticklabels(components)
ax.legend()

# Optional: Werte oberhalb der Balken anzeigen
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)

plt.tight_layout()

# Speichern der Diagramme als SVG-Dateien im Unterordner "Volllaststunden"
output_path = os.path.join(output_dir, 'grouped_bar_chart_example.svg')
plt.savefig(output_path, format='svg')
plt.show()
```
