# PEM

```python
def linearize_function(func, p_min, p_max, num_points):
    # Erzeuge p-Werte im spezifizierten Bereich
    p = np.linspace(p_min, p_max, num_points).reshape((-1, 1))
    
    # Erzeuge E_out-Werte basierend auf der Funktion
    E_out = func(p)
    
    # Erstelle das lineare Regressionsmodell
    model = LinearRegression()
    model.fit(p, E_out)
    
    # Vorhersagen
    E_out_pred = model.predict(p)
    
    # Plot der Daten und der Regressionslinie
    plt.scatter(p, E_out, color='black', label='Originaldaten')
    plt.plot(p, E_out_pred, color='blue', linewidth=3, label='Regressionslinie')
    plt.xlabel('p')
    plt.ylabel('E_out')
    plt.title('Linearisierung der Funktion')
    plt.legend()
    plt.show()
    
    # Anzeigen der Koeffizienten und des Intercepts
    print(f"Steigung (Slope): {model.coef_[0][0]}")
    print(f"y-Achsenabschnitt (Intercept): {model.intercept_[0]}")
    
    # Berechnung des mittleren quadratischen Fehlers (MSE)
    mse = mean_squared_error(E_out, E_out_pred)
    print(f"Mean Squared Error: {mse}")
    
    # Berechnung der E_out-Werte für p_min und p_max
    E_out_min = model.predict(np.array([[p_min]]))[0][0]
    E_out_max = model.predict(np.array([[p_max]]))[0][0]
    print(f"E_out-Wert für p_min ({p_min}): {E_out_min}")
    print(f"E_out-Wert für p_max ({p_max}): {E_out_max}")

# Beispiel-Funktion (kann durch jede andere Funktion ersetzt werden)
def example_function(p):
    return 0.63*900*((607.68*p**4 + 1100.79*p**3 - 52.6*p**2 + 1.49*p) / (1041.5*p**3 + 686.68*p**2 + 10.9*p - 2.33 + (0.12/p)))

# Parameter für die Analyse
p_min = 0.2
p_max = 1.2
num_points = 20

# Linearisierung der Beispiel-Funktion
linearize_function(example_function, p_min, p_max, num_points)

def E_out_new(p):
    return 
```
